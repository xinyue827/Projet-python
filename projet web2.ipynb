```python
links=[]
for p in range(1,24):
    url = f"{'https://www.hellowork.com/fr-fr/emploi/recherche.html?k=data&l=%C3%8Ele-de-France&l_autocomplete=http%3A%2F%2Fwww.rj.com%2Fcommun%2Flocalite%2Fregion%2F11&c=Alternance'}&p={p}"
    driver.get(url)
    time.sleep(5)
    my_range = [i for i in range(1, 31)]
    for i in my_range:
        xpath_expression = '/html/body/main/section/div/section/ul[1]/li[{}]/div/div[2]/div[1]/span[2]/h3/a'.format(i)
        elements=driver.find_elements(By.XPATH, xpath_expression)
        if not elements: 
            continue
        link=elements[0].get_attribute('href')
        links.append(link)
```

```python
len(links)
```

```python
emails_found=[]
for link in links:
    driver.get(link)
    content = driver.page_source 
    soup = BeautifulSoup(content, 'html.parser')
    job_description= soup.find('div', class_='tw-col-span-full lg:tw-col-span-8')
    job = job_description.get_text()
    time.sleep(5)
        
    email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    emails = re.findall(email_pattern, job_description)
    
    if emails:
        for email in emails:
            emails_found.append(email)
            
print(emails_found)

```
